```python
import numpy as np
import matplotlib.pyplot as plt

def graphene_lattice(nx,ny):
    unit_length = 1.5

    # Define basis (a1, a2) for graphene lattice
    a1 = np.array([3 * unit_length / 2,  np.sqrt(3) * unit_length / 2])
    a2 = np.array([3 * unit_length / 2, -np.sqrt(3) * unit_length / 2])

    # Two-atom basis for honeycomb
    basis = [
        np.array([0.0, 0.0]),            # Atom A
        np.array([unit_length, 0.0])               # Atom B (bond length = a)
    ]

    coords = []
    for i in range(nx):
        for j in range(ny):
            R = i * a1 + j * a2
            for b in basis:
                coords.append([R[0] + b[0], R[1] + b[1]])
    return np.array(coords)


coords = graphene_lattice(nx=10, ny=10)

coords_cut = []
for c in coords:
    if c[0] >= 10 and c[0] <= 30 and c[1] >= -10 and c[1] <= 10:
        coords_cut.append(c)
coords_cut = np.array(coords_cut)

com = np.average(coords_cut,axis=0)
coords_cut = coords_cut - com

plt.figure(figsize=(6,6))
plt.plot(coords_cut[:,0],coords_cut[:,1],'o', color = 'black')


plt.xlim(-10,10)
plt.ylim(-10,10)
```
